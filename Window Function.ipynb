```python
# window : A window function performs calculations across a set of related rows while keeping each row separate in the result.
# Structure : from pyspark.sql.window import Window
# window = Window.partitionBy("<col>",..).orderBy("<col>",..)
# DF = DF.withColumn(<row_col>, row_number(window))
    # PARTITION BY → "Split into groups."
    # ORDER BY → "Sort within each group."
```

- _row_number():_ Every row gets a unique number in sequence.
- _dense_rank():_ Same row get the same rank, but no ranks are skipped.
- _rank():_ Same row get the same rank, but the next rank is skipped.

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
emp_data = [(1,'manish',50000,'IT','m'),
(2,'vikash',60000,'sales','m'),
(3,'raushan',70000,'marketing','m'),
(4,'mukesh',80000,'IT','m'),
(5,'priti',90000,'sales','f'),
(6,'nikita',45000,'marketing','f'),
(7,'ragini',55000,'marketing','f'),
(8,'rashi',100000,'IT','f'),
(9,'aditya',65000,'IT','m'),
(10,'rahul',50000,'marketing','m'),
(11,'rakhi',50000,'IT','f'),
(12,'akhilesh',90000,'sales','m')]

emp_df = spark.createDataFrame(data = emp_data).toDF("id","name","salary","dept","gender")
emp_df.createOrReplaceTempView("employee")
```

```python
window = Window.partitionBy("dept").orderBy("salary")
emp_df.withColumn("row", row_number().over(window))\
    .withColumn("Dense_rank", dense_rank().over(window))\
    .withColumn("Rank", rank().over(window)).show(truncate=False)
```

_Aggreagte Function in window :_ sum() , Avg(), Count(), Min(), Max()
- Aggregate functions used with OVER() become window functions, allowing calculations across related rows while keeping every row in the output.

```python
%sql
select * ,
min(salary) over(partition by dept order by id) as total_dept_sal
from employee
```
