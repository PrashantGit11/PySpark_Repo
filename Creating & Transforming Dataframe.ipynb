**Transformation :**

```python
employee_df = spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema","true")\
    .load("/Volumes/workspace/mde/testdata/employee_write.csv")
employee_df.show()
```

```python
employee_df.select(col("name").alias("Emp_name")).show()
```

```python
employee_df.where((col("salary") >= 150000) & ((col("age") <= 18) | (col("gender") == "f"))).show()
```

```python
employee_df.select("*",lit("Kumar").alias("last_name")).show()
```

```python
employee_df.withColumn("full_name",concat(col("name"),lit(" Kumar"))).show()
```

```python
employee_df.withColumnRenamed("id","Emp_id").show()
```

```python
employee_df.withColumn("id",col("id").cast("string"))\
    .withColumn("salary",col("salary").cast("long"))\
    .printSchema()
```

```python
employee_df.drop("address",col("gender")).show()
```
