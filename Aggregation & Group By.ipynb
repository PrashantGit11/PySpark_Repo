**Aggregation :**


_Count :_- Count is an action as well as transformation.
-     Action : DF.count()
-     transformation : DF.select(count("<col_nm>"))

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
emp_data = [
(1,'manish',26,20000,'india','IT'),
(2,'rahul',None,40000,'germany','engineering'),
(3,'pawan',12,60000,'india','sales'),
(4,'roshini',44,None,'uk','engineering'),
(5,'raushan',35,70000,'india','sales'),
(6,None,29,200000,'uk','IT'),
(7,'adam',37,65000,'us','IT'),
(8,'chris',16,40000,'us','sales'),
(None,None,None,None,None,None),
(7,'adam',37,65000,'us','IT')
]
emp_df = spark.createDataFrame(data = emp_data).toDF("id","name","age","salary","country","dept")
```

```python
emp_df.show()
```

```python
emp_df.select(count(col("Name"))).show()
emp_df.select(count(col("*"))).show()
# Count on single column --> will skip NULL value.
# count on multiple column --> will include Null as well.
```

```python
emp_df.select(sum(col("salary")).alias("total Salary")\
            ,min(col("salary")).alias("min_sal")\
            ,max(col("salary")).alias("max_sal")\
            ,avg(col("salary")).alias("avg_sal").cast("int")).show()
```

```python
emp_df.select(sum(col("salary")).alias("total_sal"),count(col("salary")).alias("total_count"))\
    .withColumn("avg_sal",col("total_sal")/col("total_count")).show()
```

**Group By :**
- What is total salary given to it's employee ?
- What is the total salary department wise ?
- I want all column name with one extra column where total salary of each dept is mentioned?
- Calculate %age of salary given to each employeeper department ?

```python
data = [(1,'manish',50000,'IT'),
(2,'vikash',60000,'sales'),
(3,'raushan',70000,'marketing'),
(4,'mukesh',80000,'IT'),
(5,'pritam',90000,'sales'),
(6,'nikita',45000,'marketing'),
(7,'ragini',55000,'marketing'),
(8,'rakesh',100000,'IT'),
(9,'aditya',65000,'IT'),
(10,'rahul',50000,'marketing')]

employee_df = spark.createDataFrame(data = data).toDF("id","Name","Salary","Dept")
```

```python
employee_df.show()
```

```python
# Total salary department wise ?
employee_df.groupBy(col("Dept")).agg(sum("salary"),count("salary"),max("salary"),min("salary")).show()
```

```python
data1 = [
    (1, "manish", 50000, "IT", "india"),
    (2, "vikash", 60000, "sales", "us"),
    (3, "raushan", 70000, "marketing", "india"),
    (4, "mukesh", 80000, "IT", "us"),
    (5, "pritam", 90000, "sales", "india"),
    (6, "nikita", 45000, "marketing", "us"),
    (7, "ragini", 55000, "marketing", "india"),
    (8, "rakesh", 100000, "IT", "us"),
    (9, "aditya", 65000, "IT", "india"),
    (10, "rahul", 50000, "marketing", "us")
]

emp_df1 = spark.createDataFrame(data = data1).toDF("id","Name","Salary","Dept","Country")
```

```python
emp_df1.groupBy("dept","Country").agg(sum("Salary").alias("Total_sal")).sort(col("Dept").desc(),col("Country")).show()
```

```python
emp_df1.createOrReplaceTempView("employee")
agg_df = spark.sql("""
          select Dept,Country,sum(salary) from employee
          group by all
          """)
agg_df.show()
```
