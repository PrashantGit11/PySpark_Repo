```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
emp_data = [
(1,'manish',26,20000,'india','IT'),
(2,'rahul',None,40000,'germany','engineering'),
(3,'pawan',12,60000,'india','sales'),
(4,'roshini',44,None,'uk','engineering'),
(5,'raushan',35,70000,'india','sales'),
(6,None,29,200000,'uk','IT'),
(7,'adam',37,65000,'us','IT'),
(8,'chris',16,40000,'us','sales'),
(None,None,None,None,None,None),
(7,'adam',37,65000,'us','IT')
]

emp_df = spark.createDataFrame(data = emp_data).toDF("id","name","age","salary","country","dept")
```

```python
emp_df.show()
```

_case stsatement : when-otherwise_

```python
emp_df.withColumn("Adult",when(col("age") >= 18,"Yes").when(col("age") < 18, "No").otherwise("NA")).show()
# Best Practice -> Use lit() when you want to make it clear you're supplying a constant value.
```

```python
emp_df.withColumn("age",when(col("age").isNull(),lit(19)).otherwise(col("age")))\
.withColumn("Adult",when(col("age") >= 18,"Yes").when(col("age") < 18, "No").otherwise("NA")).show()
```

```python
emp_df.withColumn("Adult", when(((col("age") > 0) & (col("age") < 18)),lit("Minors"))\
                            .when(((col("age") >= 18) & (col("age") <= 30)),lit("Middle Aged"))\
                            .when((col("age")) > 30,lit("Adults"))\
                            .otherwise(lit("NA"))).show()
```

**Unique :**
- How to find unique value in spark ?
- How to drop duplicate rows ?
- How to sort data in ascending & descending order ?

```python
data=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17),
(15 ,'Mohit',45000,  18),
(13 ,'Nidhi',60000,  17),      
(14 ,'Priya',90000,  18),  
(18 ,'Sam',65000,   17)
     ]
manager_df = spark.createDataFrame(data = data).toDF("id","Name","sal","mngr_id")
```

```python
manager_df.show()
```

```python
manager_df.distinct().show()
```

```python
manager_df.select("id","Name").distinct().show()
# distinct() does not take any arguments. It removes duplicate rows by comparing all columns and returns only unique records.
```

_Dropping Duplicate :_

```python
manager_df.drop_duplicates().show()
# distinct() → removes duplicates by comparing all columns.
# drop_duplicates() -> takes subset of column(s) name as argument.
# df.distinct() is equivalent to df.dropDuplicates() with no arguments.
```

_Sorting :_

```python
manager_df.sort(col("sal").desc(),col("Name")).show()
```
