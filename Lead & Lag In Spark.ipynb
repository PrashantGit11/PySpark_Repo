**Lead & Lag :**
- What is %age of sales each month based on last 6 months sales ?
- What is %age of loss or gain based on previous month sales ?

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
product_data = [
(1,"iphone","01-01-2023",1500000),
(2,"samsung","01-01-2023",1100000),
(3,"oneplus","01-01-2023",1100000),
(1,"iphone","01-02-2023",1300000),
(2,"samsung","01-02-2023",1120000),
(3,"oneplus","01-02-2023",1120000),
(1,"iphone","01-03-2023",1600000),
(2,"samsung","01-03-2023",1080000),
(3,"oneplus","01-03-2023",1160000),
(1,"iphone","01-04-2023",1700000),
(2,"samsung","01-04-2023",1800000),
(3,"oneplus","01-04-2023",1170000),
(1,"iphone","01-05-2023",1200000),
(2,"samsung","01-05-2023",980000),
(3,"oneplus","01-05-2023",1175000),
(1,"iphone","01-06-2023",1100000),
(2,"samsung","01-06-2023",1100000),
(3,"oneplus","01-06-2023",1200000)
]
product_df = spark.createDataFrame(data = product_data).toDF("product_id","product_name","sales_date","sales") 
```

_lag() :_ LAG(column_name, offset, default_value)
- offset(optional) - How many rows back to look
- default_value(optional) - Value returned when there is no previous row.

```python
window = Window.partitionBy("product_id").orderBy("sales_date")
product_df.withColumn("previous_month_sales", lag(col("sales"),1).over(window))\
    .withColumn("%_profit_loss",round((col("sales")-col("previous_month_sales"))/(col("previous_month_sales")*0.01),2) ).show()
```

```python
product_df.show()
```

```python
# What is %age of sales each month based on last 6 months sales ?
window = Window.partitionBy("product_id")
product_df.withColumn("total_sales", sum("sales").over(window))\
    .withColumn("%_sales",round(((col("sales") /col("total_sales"))*100),2))\
        .drop("total_sales").show()
```
