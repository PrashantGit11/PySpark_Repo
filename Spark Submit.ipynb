```python
-> Spark-submit is a single command that is used to submit any spark application (Python, Scala, Java or R) to any cluster manager 
   (local, standalone, YARN, Mesos, Kubernetes)

	spark-submit --master yarn \
		--deploy-mode cluster \ (client)
        --conf spark.dynamicAllocation.enabled = true\
        --conf spark.dynamicAllocation.minExecutors = 1\
        --conf.spark.dynamicAlloaction.maxExecutors = 10\
        --conf.spark.sql.broadcastTimeout = 3600\
		--driver-memory 2G \
		--executory-memory 5G \
		--driver-cores 2 \
		--executor-cores 5 \
		--num-executors 10 \
		E:\\Spark\\wordcount.py [app args]
```

```python
+----------------+
|   Edge Node    |  (User logs into the edge node which provide user the cluster's entry point or gateway machine)
| (Client/User)  |  (Authentication and authorization can be done at this level)
+-------+--------+
        |
        | Submit Spark Job
        v
+-------------------------+
| Cluster Manager         |
| (YARN / K8s / Spark)    |
+-----------+-------------+
            |
    +-------+-------+
    |               |
    v               v
+--------+      +--------+
| Worker |      | Worker |
| Node 1 |      | Node 2 |
+--------+      +--------+
```

```python
Deploy mode : 
    Client : Driver runs on client machine(edge node).
        pros : used during development & testing. Interactive mode.
        cons : if by any reason Client machine shutoff entire application will get killed.
    Cluster : Driver runs on one of the worker node along with executor.(you get an application Id to access spark UI.)
        pros : No dependency on client machine ; no network latency
```

```python
spark-submit = Starts and deploys your Spark application on a cluster.
SparkSession.builder = Creates a Spark session inside the application so your code can interact with Spark.
                       SparkSession provides access to and uses the resources allocated by spark-submit
                        (entry point for working with DataFrames, SQL, and Spark APIs)
```
