**Union & Union ALL :**
- Difference between union & Union ALL ?
- what will happen if I change Number of columns while unioning the data ?
- What if column name is different ?
- What is unionByName ?

```python
# df1.union(df2) = SQL UNION ALL (keeps duplicates)
# df1.union(df2).distinct() = SQL UNION (removes duplicates)
```

```python
data=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17)]

schema = ("id","name","salary","mgr_id")

manager_df = spark.createDataFrame(data= data,schema = schema)
```

```python
manager_df.show()
```

```python
manager_df.count()
```

```python
data1=[(19 ,'Sohan',50000, 18),
(20 ,'Sima',75000,  17)]

schema1 = ("id","name","salary","mgr_id")

manager_df1 = spark.createDataFrame(data= data1,schema = schema1)
manager_df1.createOrReplaceTempView("manager1")
```

```python
manager_df1.count()
```

```python
manager_df.union(manager_df1).count()
```

```python
manager_df.unionAll(manager_df1).count()
```

_Duplicate :_

```python
data2=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17),
(18 ,'Sam',55000,   17),
(18 ,'Sam',65000,   17)]

schema2 = ("id","name","salary","mgr_id")

duplicate_manager_df = spark.createDataFrame(data= data2,schema = schema2)
duplicate_manager_df.createOrReplaceTempView("duplicate_manager")
```

```python
duplicate_manager_df.union(manager_df1).count()
```

```python
duplicate_manager_df.unionAll(manager_df1).count()
```

```python
spark.sql("""
select * from duplicate_manager
union
select * from manager1
""").count()
```

```python
spark.sql("""
select * from duplicate_manager
union All
select * from manager1
""").count()
```

_column Order Changed :_

```python
wrong_column_data=[(19 ,50000, 18,'Sohan'),
(20 ,75000,  17,'Sima')]

schema = ("id","salary", "mgr_id", "Name")

wrong_manager_df = spark.createDataFrame(data = wrong_column_data,schema = schema)
```

```python
manager_df1.union(wrong_manager_df).show()
```

```python
manager_df1.unionByName(wrong_manager_df).show()
```
