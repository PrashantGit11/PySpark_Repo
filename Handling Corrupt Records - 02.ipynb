**Handling Corrupt Records**
- When do you say that records are corrupted ?
- What will happen when we encounter with corrupted records in different read mode ?
- How can we print bad records ?
- Where do you store corrupted records and how can we access it later ?

```python
corrupt_emp_df = spark.read.format("csv")\
    .option("header",True) \
    .option("inferSchema",True)\
    .option("mode","DROPMALFORMED") \
    .load("/Volumes/workspace/mde/testdata/employee_file_corrupt.csv")
```

```python
corrupt_emp_df.show()
```

**Printing Bad records :** For Printing bad records needs to manually pass the schema.

```python
from pyspark.sql.types import StructType,StructField,IntegerType,StringType
my_schema = StructType([
    StructField("id",IntegerType(),True),
    StructField("name",StringType(),True),
    StructField("age",IntegerType(),True),
    StructField("salary",IntegerType(),True),
    StructField("address",StringType(),True),
    StructField("nominee",StringType(),True),
    StructField("_corrupt_record",StringType(),True)
])
```

```python
print_corrupt_emp_df = spark.read.format("csv")\
    .option("header",True) \
    .option("inferSchema",True)\
    .option("mode","PERMISSIVE") \
    .schema(my_schema)\
    .load("/Volumes/workspace/mde/testdata/employee_file_corrupt.csv")

# works only with "PERMISSIVE"
```

```python
print_corrupt_emp_df.show(truncate = False)
```

**Storing Corrupt Records :**

```python
store_corrupt_emp_df = spark.read.format("csv")\
    .option("header",True) \
    .option("badRecordsPath","/Volumes/workspace/mde/testdata/bad_records") \
    .schema(my_schema)\
    .load("/Volumes/workspace/mde/testdata/employee_file_corrupt.csv")

# mode is not allowed while storing corrupt records.
```

```python
store_corrupt_emp_df.show()
```
