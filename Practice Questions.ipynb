```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Write Query to report the name of customer that area not referred by the customer with id = 2
leet_code_data = [
    (1, 'Will', None),
    (2, 'Jane', None),
    (3, 'Alex', 2),
    (4, 'Bill', None),
    (5, 'Zack', 1),
    (6, 'Mark', 2)
]
customer_df = spark.createDataFrame(data = leet_code_data).toDF("id","name","referee_id")
```

```python
customer_df.where((col("referee_id") != 2) | (col("referee_id").isNull())).show()
# NULL is "unknown". Spark cannot say whether NULL is equal to 2 or not equal to 2, so the result is NULL, and where() only keeps TRUE conditions
```
